### Energy Formula

```julia
n=size(All_dis)[1]
function f(v,d)    #Energy Consumption Formula (22000)
    z=[]
    l=size(v)[1]
    
    for i=1:l
        push!(z,(1.2)*(1/0.5)*(0.5*1.2*0.29*2.27*v[i]^2+0.012*17758*9.81)*d[i]*(1/(3.6*10^6)))# 17758
    end
    return z
end  
energy_consumption=[]
for i=1:n
    push!(energy_consumption,f(All_speed[i],All_dis[i]))
end
```

## System Model

```julia
# I would like to track battery status based on the route
Cbat=100*30 # Battery cost per kwh
Cinv= 20000 # Inverter cost per unit
Ccab=200 # Cable cost per unit
#Ccab=200/30 # Cable cost per unit
l=6 # number of roads: we do not calculate battery cost for depot routes
k=[2,3,3,3];# number of buses on each route
#k=[2,3,4,4];
supply=80 # energy supply rate per kw
alpha=0.8
beta=0.2;
using JuMP, Gurobi, Cbc
n=length(distance)
m=Model(Gurobi.Optimizer)
#m=Model(with_optimizer(Cbc.Optimizer))
@variable(m, y[i=1:n], Bin) 
@variable(m, I[r=1:l, i=1:n]>=0) # Battery Status  
@variable(m, z[i=1:n], Bin) # inverter (0 or 1) 
@variable(m, emax[i=1:l]>=0) # For maximum battery size  
@variable(m, Ilow[i=1:l]>=0)
@variable(m, Ihigh[i=1:l]>=0)
#@objective(m, Min, Cbat*k*emax+Cinv*sum(z[i] for i=1:n)+Ccab*sum(y[i]*d[i] for i=1:n))
@objective(m, Min, Cbat*sum(k[i]*emax[i] for i=1:l-2)+Cinv*sum(z[i] for i=1:n )+Ccab*sum(y[i]*distance[i] for i=1:n ))
# Constraints 
for i=1:l
    @constraint(m,Ilow[i]==0.2*emax[i])
    @constraint(m,Ihigh[i]==0.8*emax[i])
end
#@constraint(m,emax[5]==8.545162757256023)                            
@constraint(m,y[1]==1) # Base station
@constraint(m,z[1]==1)
@constraint(m,y[404]==1) # depot
@constraint(m,z[404]==1) 
for r=1:l-1
     @constraint(m,I[r,1]==Ihigh[r]) # full charge in the first segment of the arcs
end                  
@constraint(m,I[6,404]==Ihigh[6])
# Building the route constraints 
                           
for r=1:l  # for each route  
    for i=2:length(routes_all[r]) # for each arc in the selected route
    
        @constraint(m,I[r,routes_all[r][i]]==I[r,routes_all[r][i-1]]+supply*y[routes_all[r][i]]*time[routes_all[r][i]]/(3600)-energy[routes_all[r][i]])                                  
        @constraint(m,I[r,routes_all[r][i]]<=Ihigh[r])    
        @constraint(m,I[r,routes_all[r][i]]>=Ilow[r]) 
        @constraint(m,z[routes_all[r][i]]>=y[routes_all[r][i]]-y[routes_all[r][i-1]])
           #@constraint(m,z[i,1]+z[s,size(All_dis[s])[1]]<=1)
    end
end
for i=1:l-2
     @constraint(m,emax[i]>=emax[l]) 
     @constraint(m,emax[i]>=emax[l-1]) 
end
                            

# The following condition to prevent overload for the installed inverter 

for r=1:l  # for each route  
    for i=1:length(routes_all[r])-11
        @constraint(m,sum(y[routes_all[r][k]] for k=i:i+11)<=10)
    end
end

#@constraint(m,sum(y)==10)


# How to determine number of inverters 
status =JuMP.optimize!(m)
#print(m)
println("Objective value: ", JuMP.objective_value(m))
```

```julia
# One battery V2
Cbat=100 # Battery cost per kwh
Cinv= 20000/30 # Inverter cost per unit
Ccab=200/30 # Cable cost per unit
#Ccab=200/30 # Cable cost per unit
l=6 # number of roads: we do not calculate battery cost for depot routes
k=25;# number of buses on each route
supply=80 # energy supply rate per kw
alpha=0.8
beta=0.2;
using JuMP, Gurobi, Cbc
n=size(All_dis)[1]

m=Model(with_optimizer(Gurobi.Optimizer))
#m = Model(with_optimizer(Cbc.Optimizer))
@variable(m, y[i=1:n,j=1:size(All_dis[i])[1]], Bin) 
@variable(m, I[i=1:n,j=1:size(All_dis[i])[1]]>=0) # Battery Status  
@variable(m, z[i=1:n,j=1:size(All_dis[i])[1]], Bin) # inverter (0 or 1) 
@variable(m, emax>=0) # For maximum battery size  
@variable(m, Ilow>=0)
@variable(m, Ihigh>=0)
#@objective(m, Min, Cbat*k*emax+Cinv*sum(z[i] for i=1:n)+Ccab*sum(y[i]*d[i] for i=1:n))
@objective(m, Min, Cbat*k*emax+Cinv*sum(z[i,j] for i=1:n for j=1:size(All_dis[i])[1])+Ccab*sum(y[i,j]*All_dis[i][j] for i=1:n for j=1:size(All_dis[i])[1]))
# Constraints 

@constraint(m,Ilow==0.2*emax)
@constraint(m,Ihigh==0.8*emax)


@constraint(m,y[1,1]==1) 
@constraint(m,z[1,1]==1)
@constraint(m,y[34,1]==1) # depot
@constraint(m,z[34,1]==1) 
@constraint(m,I[1,1]==Ihigh)         
@constraint(m,I[34,1]==Ihigh)
# Building the route constraints 
s=1;                            
for r=1:l  # for each route  
    for i in Route_ind[r] # for each arc in the selected route
        if i !=1 && i !=34                                      
           @constraint(m,I[i,1]<=I[s,size(All_dis[s])[1]]+supply*y[i,1]*All_time[i][1]/(3600)-energy_consumption[i][1]) 
           @constraint(m,I[i,1]<=Ihigh)
           @constraint(m,I[i,1]>=Ilow)  
           @constraint(m,z[i,1]>=y[i,1]-y[s,size(All_dis[s])[1]])
           @constraint(m,z[i,1]+z[s,size(All_dis[s])[1]]<=1)
        end            
        for j=2:size(All_dis[i])[1]                         
              @constraint(m,I[i,j]<=I[i,j-1]+supply*y[i,j]*All_time[i][j]/(3600)-energy_consumption[i][j])
              @constraint(m,I[i,j]<=Ihigh)
              @constraint(m,I[i,j]>=Ilow)            
             
        end
        s=i # keep the arc# to next arc, so we can updated first segment in arc#2 from last segment of Arc#1
    end
end
                           
for i=1:n
    for j=2:size(All_dis[i])[1]
         @constraint(m,z[i,j]+z[i,j-1]<=1)
         @constraint(m,z[i,j]>=y[i,j]-y[i,j-1])
    end
end

# The following condition to prevent overload for the installed inverter 
for i=1:n
    for j=1:size(All_dis[i])[1]-4
         @constraint(m,sum(y[i,k] for k=j:j+4)<=3)
        
    end
end

#for i=1:n-5 
    #  @constraint(m, sum(y[j] for j=i:i+5)<=5)

#end
# How to determine number of inverters 
status =JuMP.optimize!(m)
#print(m)
println("Objective value: ", JuMP.objective_value(m))
```
